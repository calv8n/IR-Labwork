```python
import math
from collections import Counter

class CustomTFIDFVectorizer:
    def __init__(self):
        self.documents = []
        self.term_idf = {}

    def add_document(self, document):
        self.documents.append(document)

    def _compute_idf(self):
        total_documents = len(self.documents)
        term_document_count = {}
        for doc in self.documents:
            terms = set(doc.split())
            for term in terms:
                if term in term_document_count:
                    term_document_count[term] += 1
                else:
                    term_document_count[term] = 1
        for term, doc_count in term_document_count.items():
            self.term_idf[term] = math.log(total_documents / doc_count)

    def transform(self, query):
        if not self.term_idf:
            self._compute_idf()

        query_terms = query.split()
        query_vector = Counter(query_terms)

        tfidf_vector = {}
        for term, idf in self.term_idf.items():
            tf = query_vector.get(term, 0) / len(query_terms)
            tfidf_vector[term] = tf * idf

        return tfidf_vector


```

```python
vectorize = CustomTFIDFVectorizer()
```

```python
for i in text:
    vectorize.add_document(i)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics.pairwise import cosine_similarity

def plot_cosine_similarity(tfidf_matrix):
    # Compute cosine similarity between documents
    cosine_sim_matrix = cosine_similarity(tfidf_matrix)

    # Create a plot for the cosine similarity matrix
    plt.figure(figsize=(8, 6))
    plt.imshow(cosine_sim_matrix, cmap='coolwarm', interpolation='nearest')
    plt.colorbar()
    plt.title('Cosine Similarity Matrix')
    plt.show()

# Assuming you have the TF-IDF matrix from the custom vectorizer
tfidf_matrix = vectorizer.transform(query)

# Plot cosine similarity
plot_cosine_similarity(tfidf_matrix)
```
